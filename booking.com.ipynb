```python
# time.sleep(1)
# query_txt = input("크롤링할 장소는?")
# destination= driver.find_element_by_css_selector('#ss')
# destination.send_keys(query_txt)
# driver.find_element_by_css_selector('.sb-searchbox__button').click()

# driver.find_element_by_css_selector('label.a1b3f50dcd.b2fe1a41c3.a1f3ecff04.d19ba76520.efeda70352>span.bbdb949247').click()

page = 1
hotel_name = []
distant = []
cost = []
rate = []
image = []
sort = []

url = []
address = []
desc = []

kind = []
clean = []
facility  = []
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import NoSuchElementException

page = 1
while page<2:
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    driver.implicitly_wait(10)
    ## 이름
    name = driver.find_elements_by_css_selector("div.dd023375f5>h3.a4225678b2")
    for i in range(len(name)):
        call = name[i].text
        hotel_name.append(call)


    
    ## 거리
    distance = driver.find_elements_by_css_selector('[data-testid="distance"]')
    for i in range(len(distance)):
        distant.append(distance[i].text)


    ## 가격
    price = driver.find_elements_by_css_selector("span.fcab3ed991.bd73d13072")
    for i in range(len(price)):
        cost.append(price[i].text)

 
    ## 평점
    rating = driver.find_elements_by_css_selector("div.b5cd09854e.d10a6220b4")
    for i in range(len(rating)):
        rate.append(rating[i].text)

    # 이미지
    imgs = driver.find_elements_by_css_selector('img.b8b0793b0e')
    for img in imgs:                                #모든 이미지들을 탐색
        image.append(img.get_attribute('src'))    
    
    ## 상세페이지
    list = driver.find_elements_by_css_selector('div.a826ba81c4.fe821aea6c.fa2f36ad22.afd256fc79.d08f526e0d.ed11e24d01.da89aeb942')
    
    for data in list:
        data.find_element_by_css_selector('a.e13098a59f').click()
        driver.switch_to.window(driver.window_handles[1])
        
        time.sleep(1)
        driver.implicitly_wait(10)
        ## 구분
        sorting = driver.find_element_by_css_selector('h2.hp__hotel-name>span')
        sort.append(sorting.text)

        ## 주소
        addresses = driver.find_element_by_css_selector("span.hp_address_subtitle.js-hp_address_subtitle.jq_tooltip")
        address.append(addresses.text)
    

        ##설명
        descs = driver.find_element_by_id("property_description_content")
        desc.append(descs.text)

        ## 세부 점수
#         elements = driver.find_elements_by_css_selector('div.v2_review-scores__body.v2_review-scores__body--compared_to_average> ul > li')

        url.append(driver.current_url)

        try: 
            kinde = driver.find_element_by_css_selector('#guest-featured_reviews__horizontal-block > div > div.review_list_score_container.hp_subscore_explanation_container > div > div > ul > li:nth-child(1) > div > span.c-score-bar__score')
            cleane = driver.find_element_by_css_selector('#guest-featured_reviews__horizontal-block > div > div.review_list_score_container.hp_subscore_explanation_container > div > div > ul > li:nth-child(3) > div > span.c-score-bar__score')
            facilitye = driver.find_element_by_css_selector('#guest-featured_reviews__horizontal-block > div > div.review_list_score_container.hp_subscore_explanation_container > div > div > ul > li:nth-child(5) > div > span.c-score-bar__score')
            
            kind.append(kinde.text)
            clean.append(cleane.text)
            facility.append(facilitye.text)

        except NoSuchElementException as n: 
            kind.append('0')
            clean.append('0')
            facility.append('0')
            

        
            
        ## 수집 종료
        driver.close()
        driver.switch_to.window(driver.window_handles[0])

    driver.implicitly_wait(10)
    
    try:
        driver.find_element_by_css_selector('div.f32a99c8d1.f78c3700d2').click()
    except:
        break
    time.sleep(3)
    
    page = page+1
```
